### HAndling outliers (5 number summary and box plot)

##### 5 Number Summary : min,max,q1,q2,q3

```python
import numpy as np
lst_marks=[45,32,56,75,89,54,32,89,90,87,67,54,45,98,99,67,74]
min,q1,q2,q3,max=np.quantile(lst_marks,[0,0.25,0.5,0.75,1])
min,q1,q2,q3,max
```

```python
IQR=q3-q1
print(IQR)
```

```python
lower_fence=q1-1.5*(IQR)
higher_fence=q3+1.5*(IQR)
```

```python
lower_fence,higher_fence
```

anything outside the (lower fence,higher fence) is considered as an outlier

```python
import seaborn as sns
sns.boxplot(lst_marks)
```

now lets add some outliers

```python
lst_marks=[-100,-200,45,32,56,75,89,54,32,89,90,87,67,54,45,98,99,67,74,150,170,180]
sns.boxplot(lst_marks)
```
