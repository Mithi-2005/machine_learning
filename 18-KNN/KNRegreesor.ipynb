### K nearest Neighnour Regreessor problem

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from sklearn.datasets import make_regression
x,y=make_regression(n_samples=1000,n_features=3,noise=10,random_state=42)
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsRegressor
regressor=KNeighborsRegressor(
    n_neighbors=6, ## K value
    algorithm='auto' ,## Alogrithm KDTree , Ball Tree etc 
    p=2, ##Euclidian or manhattan Distance
)
```

```python
regressor.fit(x_train,y_train)
```

```python
from sklearn.metrics import r2_score
y_pred=regressor.predict(x_test)
score=r2_score(y_test,y_pred)
print("Accuracy Score: ",score)
```
