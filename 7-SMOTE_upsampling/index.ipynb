### SMOTE Upsampling


```python
from sklearn.datasets import make_classification
```

```python
x,y=make_classification(
    n_samples=1000, #total samples
    n_features=2, #Fratures count
    n_clusters_per_class=1, #Groups per class
    weights=[0.90], #ratio
    n_redundant=0, #DOnt want this so set to 0 (Default is 2)
    random_state=123 #Random seed
) #IT will return two arrays as a tuple(fetaures array,target array)
```

```python
import pandas as pd
df1=pd.DataFrame(x,columns=['feature 1','feature 2'])
df2=pd.DataFrame(y,columns=['target'])
df=pd.concat([df1,df2],axis=1)
df['target'].value_counts()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['feature 1'],df['feature 2'],c=df['target'])
```

# Insall smote using pip insall imblearn

```python
from imblearn.over_sampling import SMOTE # type: ignore
```

```python
oversample=SMOTE()
x.shape
```

```python
x,y=oversample.fit_resample(df[['feature 1','feature 2']],df['target'])
x.shape
```

```python
df1=pd.DataFrame(x,columns=['feature 1','feature 2'])
df2=pd.DataFrame(y,columns=['target'])
df=pd.concat([df1,df2],axis=1)
df
```

```python
plt.scatter(df['feature 1'],df['feature 2'],c=df['target'])
```
