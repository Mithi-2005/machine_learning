```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_circles ##Dataset in moon shape
```

```python
x,y=make_circles(n_samples=750,noise=0.1,factor=0.3)
```

```python
x
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
```

```python
from sklearn.cluster import DBSCAN
dbsan=DBSCAN(eps=0.1)
```

```python
pred=dbsan.fit_predict(x) ## ALL -1 values are outliers
idx=np.where(pred<0)
```

```python
plt.scatter(x[:,0],x[:,1],c=pred)
plt.scatter(x[idx,0],x[idx,1],edgecolors='red')
```
