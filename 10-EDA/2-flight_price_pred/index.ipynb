### EDA On Flight Prediction Dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_excel('flight_price.xlsx')
df.head()
```

```python
df.info()
```

```python
df['Date']=df['Date_of_Journey'].str.split("/").str[0]
df['Month']=df['Date_of_Journey'].str.split("/").str[1]
df['Year']=df['Date_of_Journey'].str.split("/").str[2]
df
```

```python
df.info() #Still date onth and year are object type
```

```python
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
df.info()
```

```python
# SO we can drop Date_of_Journey
df.drop(columns=['Date_of_Journey'],axis=1,inplace=True)
```

```python
df
```

#### Now lets work with arrival time

```python
df['Arrival_hour']=df['Arrival_Time'].str.split(':').str[0]
df['Arrival_min']=df['Arrival_Time'].str.split(':').str[1]
```

```python
df['Arrival_min']=df['Arrival_min'].str.split().str[0]
```

```python
df
```

```python
df.info() # Now convert Arrival hour and min column into int data type
```

```python
df['Arrival_hour']=df['Arrival_hour'].astype(int)
df['Arrival_min']=df['Arrival_min'].astype(int)
df.info()
```

```python
df.drop(columns=['Arrival_Time'],axis=1,inplace=True)
```

```python
df['Departure_hour']=df['Dep_Time'].str.split(':').str[0]
df['Departure_min']=df['Dep_Time'].str.split(':').str[1]
df

```

```python
df['Departure_hour']=df['Departure_hour'].astype(int)
df['Departure_min']=df['Departure_min'].astype(int)
df.info()
```

```python
df.drop(columns=['Dep_Time'],axis=1,inplace=True)
```

```python
df.info()
```

#### LEts go with duration

```python
df[['dur_hr','dur_min']]=df['Duration'].str.split(' ',expand=True)
df
```

```python
df['dur_hr'].value_counts()
```

```python
def correct_duration(row):
    if 'm' in str(row['dur_hr']):
        row['dur_min'] = row['dur_hr']
        row['dur_hr'] = '0h'
    return row

df = df.apply(correct_duration, axis=1)
```

```python
df['dur_hr'].value_counts()
```

```python
df['dur_min']=df['dur_min'].fillna("0m")
```

```python
df['dur_hr']=df['dur_hr'].astype(str).str.replace('h','').astype(int)
df['dur_min']=df['dur_min'].astype(str).str.replace('m','').astype(int)
df
```

```python
df['Duration']=df['dur_hr']*60+df['dur_min']
df.drop(columns=['dur_hr','dur_min'],axis=1,inplace=True)
```

```python
df.info()
```

#### Now lets go with total stops

```python
df['Total_Stops'].unique()
```

```python
df['Total_Stops'].isnull().sum()
```

```python
df['Total_Stops'].mode()
```

```python
##We can replace that only NaN with mode
df['Total_Stops']=df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,np.nan:1})
```

```python
df['Total_Stops']=df['Total_Stops'].fillna(1)
```

```python
df['Total_Stops']=df['Total_Stops'].astype(int)
df.head()
```

```python
## I dont want route because i have source and dest i will drop it
df.drop('Route',axis=1,inplace=True)
```

```python
df.info()
```
