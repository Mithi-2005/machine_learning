### Exploratory Data Analysis(EDA) ON RED WHINE DATASET

```python
import  pandas as pd
df=pd.read_csv('winequality-red.csv')
df.head()
```

```python
# Gives summary of the dataset
df.info()
```

```python
#Gives descriptive stats of the dataset like mean,var,SD etc
df.describe()
df.shape # Give columns and rows
df.columns #lists all the columns
```

```python
# Gets all the unique values in the columns
df['quality'].unique()
```

```python
#GEts the count of missing values
df.isnull().sum()
```

```python
#Find the duplicate records
df[df.duplicated()==True] #Total 240 duplicates
df.drop_duplicates() #drops all duplicates 
```

```python
## Get the correlation of each feature with respect to each feature
df.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(),annot=True)
```

```python
df['quality'].value_counts()
#conclusion is it is a imbalanced dataset
df['quality'].value_counts().plot(kind='bar')
```

```python
sns.histplot(df['alcohol'],kde=True) #Gives graph that can help to recognise distribution
```

```python
#univariate,bivariate,multivariate analysis
sns.pairplot(df)
```

```python
#Categorical Plot
sns.catplot(x='quality',y='alcohol',data=df,kind='box')
```

```python
sns.scatterplot(x='alcohol',y='pH',data=df,hue='quality')
```
