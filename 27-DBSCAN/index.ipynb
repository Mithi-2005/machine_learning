#### DBSCAN

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_moons ##Dataset in moon shape
```

```python
x,y=make_moons(n_samples=250,noise=0.05)
x
```

```python
plt.scatter(x[:,0],x[:,1],c=y) ###Moon shapes
```

```python
###Standard Scaling
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
```

```python
from sklearn.cluster import DBSCAN
dbscan=DBSCAN(eps=0.3)
```

```python
dbscan.fit(x_scaled)
```

```python
dbscan.labels_
```

```python
plt.scatter(x[:,0],x[:,1],c=dbscan.labels_) ###Moon shapes
```
